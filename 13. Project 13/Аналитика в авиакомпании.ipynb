# Спрос на рейсы авиакомпании

Вы аналитик российской авиакомпании F9, выполняющей внутренние пассажирские перевозки. Важно понять предпочтения пользователей, покупающих билеты на разные направления.

Вам предстоит изучить базу данных и проанализировать спрос пассажиров на рейсы в города, где проходят крупнейшие культурные фестивали.

У вас есть файлы с результатами запросов из предыдущих заданий.

**query_1.csv** — результат первого запроса. В нём содержится такая информация:

- *model* — **модель самолета;
- *flights_amount* — количество рейсов для каждой модели самолетов *model* в сентябре 2018 года.

**query_3.csv** — результат третьего запроса. В нём содержится такая информация:

- *city* — город;
- *average_flights* — среднее количество рейсов, прибывающих в город (*city*) за день в сентябре 2018 года.

 Для этих двух наборов данных нужно:

- импортировать файлы;
- изучить данные в них;
- проверить типы данных на корректность;
- выбрать топ-10 городов по количеству рейсов;
- построить графики: модели самолётов и количество рейсов, города и количество рейсов, топ-10 городов и количество рейсов;
- сделать выводы по каждому из графиков, пояснить результат.


## Аналитика

### Импортируем данные

```python
#Загрузим библиотеки
import pandas as pd
import matplotlib.pyplot as pl
import matplotlib.pyplot as plt

#Импортируем данные
data_1 = pd.read_csv('/datasets/query_1.csv')
data_3 = pd.read_csv('/datasets/query_3.csv')
```

### Изучим данные

```python
#Отобразим первые 10 строк каждого Датасета
display(data_1.head(10))
display(data_3.head(5))
```

### Проверим типы данных на корректность

```python
#Отобразим информацию о Датасетах
print('Информация о датасете data_1:\n')
data_1.info()
print('\nИнформация о датасете data_3:\n')
data_3.info()
```

### Выберим топ 10 городов по количеству рейсов

```python
top10_city = data_3.sort_values(by='average_flights', ascending=False).head(10)
display(top10_city)
```

### Построим график зависимости модели самолётов и количество рейсов

```python
data_1_sort = data_1.sort_values(by='flights_amount', ascending=False)
data_1_sort.plot(y='flights_amount', x='model', figsize =(14, 6), kind = 'barh', legend = None)
plt.xlabel('Количество рейсов', fontweight='bold', fontsize=12)
plt.ylabel('Модели самолетов', fontweight='bold', fontsize=12)
plt.title('Модели самолётов и количество рейсов\n', fontweight='bold', fontsize=16)
plt.show()
```

### Вывод:

Так как в проекте рассматриваются региональные перелеты в рамках страны, то самыми используемыми самолетами для перелетов стали российские самолеты Sukhoi SuperJet-100, небольшой самолет для регионального использования Cessna 208 Caravan и маленький самолет Bombardier CRJ-200 для использования в регионах, где нет больших аэропортов (например на Чукотке).

### Построим график зависимости города и количество рейсов

```python
data_3.plot(x='city', y='average_flights', figsize =(24, 8), legend = None, kind = 'bar')
plt.xlabel('Город', fontweight='bold', fontsize=12)
plt.ylabel('Количество рейсов', fontweight='bold', fontsize=12)
plt.title('Города и количество рейсов\n', fontweight='bold', fontsize=16)
plt.show()
```

### Вывод

График подтверждает, что Москва является самым крупным городом РФ и крупнейшим авиаузлом страны (здесь больше всего аэропортов и большая чать рейсов только через Москву). Дальше идут города миллионики: Санкт-Петербург, Красноярск, Новосибирск.

### Построим график зависимости топ-10 городов и количество рейсов

```python
top10_city.plot(x='city', y='average_flights', figsize =(14, 11), legend = None, kind = 'bar')
plt.xlabel('Город', fontweight='bold', fontsize=12)
plt.ylabel('Количество рейсов', fontweight='bold', fontsize=12)
plt.title('Топ-10 городов и количество рейсов\n', fontweight='bold', fontsize=16)
plt.show()
```

### Вывод

Москва самый большой город в РФ с множеством аэропортов и частью рейсов, которые осуществляются только через Москву. Дальше иду города миллионики с крупными аэропортами. И небольшие города как Сочи куда прилетают много туристов для отдыха.
